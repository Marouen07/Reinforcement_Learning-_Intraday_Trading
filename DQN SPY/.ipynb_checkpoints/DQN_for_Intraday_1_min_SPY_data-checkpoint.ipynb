```python
def get_df(file_path):
  df=pd.read_csv(file_path,parse_dates=True) 
  df=df[['<DATE>','<TIME>','<HIGH>','<LOW>','<CLOSE>','<VOL>']]
  columns=['date','datetime','High','Low','Price','Volume']
  df.columns=columns
  df['datetime'].replace(0,np.nan,inplace=True)
  df=df.dropna()
  df['datetime']=pd.to_datetime((df['date'].astype(str)+(df['datetime']).astype(int).astype(str)),format='%Y%m%d%H%M%S')
  df['date']=pd.to_datetime(df['date'],format='%Y%m%d')
  return df 
```

### Moving Average Convergence Divergence (MACD)

Moving Average Convergence Divergence (MACD) is a trend following indicator. MACD can be calculated very simply by subtracting the 26 period EMA from the 12 period EMA. We previously discussed EMAs in our article here. MACD can be used and interpreted in a handful of different ways to give the trader potential value and insight into their trading decisions.

```python
#MACD
def MACD_df(df):
  exp1 = df.groupby('date').apply(lambda x :x['Price'].ewm(span=12, adjust=False).mean())
  exp2 = df.groupby('date').apply(lambda x :x['Price'].ewm(span=26, adjust=False).mean())
  macd = exp1-exp2
  return macd
```

### Relative Strength Index (RSI)

The Relative Strength Index (RSI) is a momentum indicator developed by noted technical analyst Welles Wilder, that compares the magnitude of recent gains and losses over a specified time period to measure speed and change of price movements of a security. It is primarily used to identify overbought or oversold conditions in the trading of an asset.

The Relative Strength Index (RSI) is calculated as follows:

    RSI = 100 - 100 / (1 + RS)

    RS = Average gain of last 14 trading days / Average loss of last 14 trading days

```python
def RSI(series, period):
    delta = series.diff()
    u = delta * 0
    d = u.copy()
    u[delta > 0] = delta[delta > 0]
    d[delta < 0] = -delta[delta < 0]
    rs = u.rolling(period).mean() / \
         d.rolling(period).mean()
    return 100 - 100 / (1 + rs)
```

```python
def RSI_df(df,period=14):
  rsi=df.groupby('date').apply(lambda x :RSI(x['Price'],period))
  return rsi
```

### Williams % R 

Williams %R, also known as the Williams Percent Range, is a type of momentum indicator that moves between 0 and -100 and measures overbought and oversold levels. The Williams %R may be used to find entry and exit points in the market. 

%R = (Highest High - Close)/(Highest High - Lowest Low) 
Lowest Low = lowest low for the look-back period
Highest High = highest high for the look-back period
%R is multiplied by -100 correct the inversion and move the decimal.

```python
def williams_R(x,period):
  close=x['Price']
  high=x['High'].rolling(period).max()
  low=x['Low'].rolling(period).min()
  williams=(high-close)/(high-low)
  return williams 
```

```python
def Williams_df(df,period=14):
  williams_ratio=df.groupby('date').apply(lambda x:williams_R(x,period)) 
  return williams_ratio
```

### On Balance Volume (OBV)
Volume is added (starting with an arbitrary number) when the market finishes higher, or volume is subtracted when the market finishes lower. This provides a running total and shows which stocks are being accumulated. It can also show divergences, such as when a price rises, but volume is increasing at a slower rate or even beginning to fall.



```python
def Preprocess_OBV(df,window):
  OBV=df['Volume']*((df['Price']>df['Price'].shift(window))*2-1)
  OBV[:window]=df['Volume'][:window]*((df['Price'][:window]>=df['Price'][0])*2-1)
  OBV.name='OBV'
  return OBV
```

```python
def OBV_df(df,window=10):
  OBV=df.groupby('date').apply(lambda x:Preprocess_OBV(x,window))
  OBV=OBV.cumsum()
  return OBV
```

```python
def feat_seq(data,feature,lookback):
  day_data=data[feature]
  sequences=[np.nan]*(lookback-1)
  for t in range(lookback,len(day_data)+1):
    if (day_data.index[t-1]-day_data.index[t-lookback]).seconds <= 60*(lookback):
      sequences+=[day_data[t-lookback:t].values]
    else : 
      sequences+=[np.nan]
  sequences=np.array(sequences)
  return sequences

def feats_seq(data,features,lookback):
  seq=[]
  for feature in features :
    seq+=[feat_seq(data,feature,lookback)]
  seq=pd.DataFrame(np.vstack(seq).T,index=data.index,columns=features)
  return seq
```

```python
import math
import random
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from collections import namedtuple
from itertools import count
from PIL import Image

import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
import torchvision.transforms as T




# if gpu is to be used
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

## Environement

```python
from collections import defaultdict
class environment(object):
  def  __init__(self,data,TC=1e-3):
    """
    data (DataFrame): Intraday Dataset 
    lookback(int): Sequence length to be fed into the LSTM
    TC: Transaction Costs (10 bps default)
    """
    self.TC=TC #Transaction costs 
    self.data=data.set_index('datetime')
    self.reset()
  def reset(self):
    self.timestamp=self.data.index[0]
    self.time_index=0
    #First state (sequence lenghth is 1 )
    self.state=np.vstack(self.data.iloc[self.time_index,1:-1].values).T
    self.position=np.array([0,0,0],dtype=np.double)
    self.day_pnl=0 # For Tracking PNL in $$$ 
    self.log_returns=0 #for Tracking Returns of Strategy
    self.carried_pnl=0
    self.orders=defaultdict(int)
  def act(self,action):
    """Take an action: 0 : hold, 1 : buy , 2 : sell """
    next_timestamp=self.data.index[self.time_index+1]
    next_state=np.vstack(self.data.iloc[self.time_index+1,1:-1].values).T

    #End of Trading Day signal 
    done=next_timestamp.date()>self.timestamp.date() 
    hold=False 

    if done :  
      #Trading Day is finished
      self.position=np.array([0,0,0],dtype=np.double)
      self.day_pnl=0 
      self.carried_pnl=0
      pnl=0
    else : 
      #Computing Environment Reward 
      prev_price=self.data.loc[self.timestamp][-1]
      new_price=self.data.loc[next_timestamp][-1]
      diff=new_price-prev_price
      if action == 1 : 
        #buy order 
        if self.position[0]==1:
          hold=True
        else:
          log_ret_TC=np.log((new_price-self.TC*(1+self.position[1]))/prev_price)
          self.position[0]=1  
          self.position[1]=0
          
          pnl=(diff-self.TC)

          self.day_pnl+=pnl
          self.carried_pnl=log_ret_TC

          #Tracking executed Orders
          self.orders[1]+=1

          #Tracking Logreturns
          self.log_returns+=log_ret_TC
      elif action == 2 :
        #sell order 
        if self.position[1]==1 :
          hold=True       
        else:
          log_ret_TC=np.log((prev_price-self.TC*(1+self.position[0]))/new_price)           
          self.position[0]=0 
          self.position[1]=1
          log_ret_TC=np.log((prev_price-self.TC)/new_price) 
          pnl=(-diff-self.TC) 

          self.day_pnl+=pnl
          self.carried_pnl=log_ret_TC

          #Tracking executed Orders
          self.orders[2]+=1

          #Tracking Logreturns
          self.log_returns+=log_ret_TC
      if action == 0 or hold == True : 
        log_ret=np.log(new_price/prev_price)
        #hold order 
        pnl=self.position[0]*diff+self.position[1]*(-diff)

        self.day_pnl+=pnl
        self.carried_pnl+=self.position[0]*log_ret+self.position[1]*-log_ret

        #Tracking executed Orders
        self.orders[0]+=1

        #Tracking Logreturns
        self.log_returns+=self.position[0]*log_ret+self.position[1]*-log_ret

      self.position[2]=pnl
    self.state=next_state
    self.timestamp=next_timestamp   
    self.time_index+=1
    return next_state,self.position,(np.exp(self.carried_pnl)-1)*100,done
```

Replay Memory
-------------

We'll be using experience replay memory for training our DQN. It stores
the transitions that the agent observes, allowing us to reuse this data
later. By sampling from it randomly, the transitions that build up a
batch are decorrelated. It has been shown that this greatly stabilizes
and improves the DQN training procedure.




The states are seperated to 2 components 

1.   Position State – this is a 3D vector of the form [𝐿, 𝑆, 𝑃𝑛𝐿] where 𝐿 represents the number
of contracts currently bought (number of long contracts in our holdings,) 𝑆 represents the
number of contracts currently sold (number of short contracts in our holdings) and 𝑃𝑛𝐿
represents the corresponding profit or loss based on the current position.

2.   Market state : we utilize the  technical indicators, along with the timestamp (encoded
to capture date-time), defined above to represent the market features.
These Features will be fed to an LSTM, hence we need to keep a sequence of these data in our memory buffer.


```python
Transition = namedtuple('Transition',
                        ('state','position', 'action', 'next_state','next_position', 'reward'))

class ReplayMemory(object):

    def __init__(self, capacity):
        self.capacity = capacity
        self.memory = list()

    def push(self, *args):
        """Saves a transition."""
        if len(self.memory) > self.capacity:
          #Memory is full, we remove first element
          del self.memory[0] 
        self.memory.append(Transition(*args))
    def flush(self):
      self.memory=list()
    def sample(self, batch_size):
        """We Sample a (Time) Sequence of states and the corresponding reward"""
        #return random.sample(self.memory, batch_size)
        return self.memory[-batch_size:]

    def __len__(self):
        return len(self.memory)
```

Now, let's define our model. But first, let quickly recap what a DQN is.

DQN algorithm
-------------

Our environment is deterministic, so all equations presented here are
also formulated deterministically for the sake of simplicity. In the
reinforcement learning literature, they would also contain expectations
over stochastic transitions in the environment.

Our aim will be to train a policy that tries to maximize the discounted,
cumulative reward
$R_{t_0} = \sum_{t=t_0}^{\infty} \gamma^{t - t_0} r_t$, where
$R_{t_0}$ is also known as the *return*. The discount,
$\gamma$, should be a constant between $0$ and $1$
that ensures the sum converges. It makes rewards from the uncertain far
future less important for our agent than the ones in the near future
that it can be fairly confident about.

The main idea behind Q-learning is that if we had a function
$Q^*: State \times Action \rightarrow \mathbb{R}$, that could tell
us what our return would be, if we were to take an action in a given
state, then we could easily construct a policy that maximizes our
rewards:

\begin{align}\pi^*(s) = \arg\!\max_a \ Q^*(s, a)\end{align}

However, we don't know everything about the world, so we don't have
access to $Q^*$. But, since neural networks are universal function
approximators, we can simply create one and train it to resemble
$Q^*$.

For our training update rule, we'll use a fact that every $Q$
function for some policy obeys the Bellman equation:

\begin{align}Q^{\pi}(s, a) = r + \gamma Q^{\pi}(s', \pi(s'))\end{align}

The difference between the two sides of the equality is known as the
temporal difference error, $\delta$:

\begin{align}\delta = Q(s, a) - (r + \gamma \max_a Q(s', a))\end{align}


```python
class DQN(nn.Module):

    def __init__(self,hidden_size ,num_layers,hidden_linear):
        super(DQN, self).__init__()
        self.bidirectional=True
        self.n_hidden=hidden_size
        self.n_layers=num_layers

        self.lstm=nn.LSTM(5,hidden_size,num_layers,batch_first=True,bidirectional=self.bidirectional)
        self.fc1=nn.Sequential(nn.Linear(3,hidden_linear),nn.ReLU())
        lstm_len=(1+self.bidirectional)*hidden_size
        self.fc2=nn.Sequential(nn.Linear(hidden_linear+lstm_len,32),nn.ReLU(),
                               nn.Linear(32,16),nn.ReLU(),nn.Linear(16,3))
        
    def init_hidden(self, batch_size,device):
        # even with batch_first = True this remains same as docs
        hidden_state = torch.zeros(self.n_layers*(1+self.bidirectional),batch_size,self.n_hidden).double()
        cell_state = torch.zeros(self.n_layers*(1+self.bidirectional),batch_size,self.n_hidden).double()
        self.hidden = (hidden_state.to(device), cell_state.to(device))
    
    def forward(self, states,positions):
      batch_size=states.size(0)
      lstm_out, self.hidden = self.lstm(states,self.hidden)
      x1 = lstm_out.contiguous()
      x1=F.relu(x1[:,-1,:]).reshape(batch_size,-1)
      x2=self.fc1(positions)
      x=torch.cat((x1,x2),dim=1)
      x=self.fc2(x)
      #x=self.fc2(x1)
      return x 
```

## Training
--------

Hyperparameters and utilities
^^^^^^^^^^^^^^^^^^^^^^^^^^^^^
This cell instantiates our model and its optimizer, and defines some
utilities:

-  ``select_action`` - will select an action accordingly to an epsilon
   greedy policy. Simply put, we'll sometimes use our model for choosing
   the action, and sometimes we'll just sample one uniformly. The
   probability of choosing a random action will start at ``EPS_START``
   and will decay exponentially towards ``EPS_END``. ``EPS_DECAY``
   controls the rate of the decay.

```python
BATCH_SIZE = 128
GAMMA = 0.999
EPS_START = 0.5
EPS_END = 0.01
EPS_DECAY = 200
TARGET_UPDATE = 10
#lookback=10
hidden_size ,num_layers,hidden_linear = 128,3,32

policy_net = DQN(hidden_size ,num_layers,hidden_linear).to(device)
policy_net=policy_net.double()
target_net = DQN(hidden_size ,num_layers,hidden_linear).to(device)
target_net=target_net.double()
target_net.load_state_dict(policy_net.state_dict())
target_net.eval()

optimizer = optim.Adam(policy_net.parameters(),lr=1e-3)
memory = ReplayMemory(264)

steps_done = 0
def select_action(state,position):
    global steps_done
    sample = random.random()
    eps_threshold = EPS_END + (EPS_START - EPS_END) * \
        math.exp(-1. * steps_done / EPS_DECAY)
    steps_done += 1
    if sample > eps_threshold:
        policy_net.init_hidden(state.shape[0],device)
        with torch.no_grad():
          policy_net.eval()
          return policy_net(state,position).max(1)[1].view(1, 1)
    else:
        return torch.tensor([[random.randrange(3)]], device=device, dtype=torch.long)

```

Training loop
^^^^^^^^^^^^^

Finally, the code for training our model.

Here, you can find an ``optimize_model`` function that performs a
single step of the optimization. It first samples a batch, concatenates
all the tensors into a single one, computes $Q(s_t, a_t)$ and
$V(s_{t+1}) = \max_a Q(s_{t+1}, a)$, and combines them into our
loss. We  use a target network to compute $V(s_{t+1})$ for
added stability. The target network has its weights kept frozen most of
the time, but is updated with the policy network's weights every so often.
This is usually a set number of steps .




```python
def optimize_model(iterations):

    if len(memory) < BATCH_SIZE:
        return
    for i in range(iterations) :
      transitions = memory.sample(BATCH_SIZE)
      batch = Transition(*zip(*transitions))

      state_batch = torch.cat(batch.state).to(device)
      position_batch=torch.cat(batch.position).to(device)
      action_batch = torch.cat(batch.action).to(device)
      reward_batch = torch.cat(batch.reward).to(device)
      next_state_batch=torch.cat(batch.next_state).to(device)
      next_position_batch=torch.cat(batch.next_position).to(device)
      batch_size=state_batch.shape[0]
      policy_net.init_hidden(batch_size,device)
      target_net.init_hidden(batch_size,device)
      # Compute Q(s_t, a) - the model computes Q(s_t), then we select the
      # columns of actions taken 
      policy_net.train() 
      state_action_values = policy_net(state_batch,position_batch).gather(1, action_batch)
      # Compute V(s_{t+1}) for all next states.
      target_net.eval()
      next_state_values= target_net(next_state_batch,next_position_batch).detach().max(1)[0].unsqueeze(1)
      #print(state_action_values.shape,next_state_values.shape,reward_batch.shape)
      # Compute the expected Q values
      expected_state_action_values = (next_state_values * GAMMA) + reward_batch
      #expected_state_action_values=expected_state_action_values.clamp(-1,1)
      # Compute Huber loss
      loss = F.mse_loss(state_action_values, expected_state_action_values)

      #loss = F.smooth_l1_loss(state_action_values, expected_state_action_values)

      # Optimize the model
      optimizer.zero_grad()
      loss.backward()
      
      """for param in policy_net.parameters():
          param.grad.data.clamp_(-5, 5)"""

      optimizer.step()
```

Below, you can find the main training loop. At the beginning we reset
the environment and initialize the ``state`` Tensor. Then, we sample
an action, execute it, observe the next screen and the reward (always
1), and optimize our model once. When the episode ends (our model
fails), we restart the loop.

Below, `num_episodes` is set small. You should download
the notebook and run lot more epsiodes.

```python
TC= 2.75* 1e-3 #0.00275 max exchange fee for NYSE #0.1 bps Transaction cost 0.003 $ approx
env=environment(data_train,TC) 
env.reset() 
memory.flush() 
```

```python
num_episodes = len(data_train)-4000
iterations=1
cum_pnl=[0]
day_pnl=[]
returns=[]
memory.flush() 
day=0
for episode in range(num_episodes):

    state=env.state 
    position=env.position 
    
    state=state.astype(float) 
    state=torch.from_numpy(state).double().unsqueeze(0) 

    position=position.astype(float) 
    position=torch.from_numpy(position).double().unsqueeze(0) 

    action = select_action(state.to(device),position.to(device)).cpu() 
    
    #We need to compute the Real PNL of The Agent 
    
 
    next_state,next_position,carried_pnl,done=env.act(action) 
      
    
    cum_pnl+=[env.day_pnl]
    returns+=[np.exp(env.log_returns)-1]
    
    next_state=next_state.astype(float)
    next_state=torch.from_numpy(next_state).double().unsqueeze(0)
    next_position=next_position.astype(float)
    next_position=torch.from_numpy(next_position).double().unsqueeze(0)
    carried_pnl=torch.Tensor([carried_pnl]).double().unsqueeze(0)  
    if not done:
      #Save Transition in Replay Buffer  
      memory.push(state,position,action,next_state,next_position,carried_pnl)
    else :
      day+=1
      day_pnl+=[cum_pnl[-2]]
      print('Money Won on day {:d} : {:.2f} $ '.format(day,cum_pnl[-2]))
      
    # Perform one step of the optimization (on the Policy network)
    optimize_model(iterations)

    # Update the target network
    if episode % (TARGET_UPDATE//iterations) == 0:
        target_net.load_state_dict(policy_net.state_dict())

    if episode % 50 == 0  :
      print("episode {:03d}/{:03d} |  Real Money Won : {:.2f} $ | immediate PNL {:.6f} | Returns : {:.3%} "
        .format(episode, num_episodes, (cum_pnl[-1]),next_position.numpy().ravel()[-1],returns[-1]))
      print(carried_pnl)
print('Complete')
```
